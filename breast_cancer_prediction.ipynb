```python
# Load data on dataframe
df = pd.read_csv('/content/data.csv')
```

```python
# count number of null(empty) values
df.isna().sum()
```

```python
# drop the column with the null values
df.dropna(axis=1, inplace=True)
```

```python
# Get count of number of M or B cells in diagnosis
df['diagnosis'].value_counts()
```

```python
# Get datatypes for each colums
df.dtypes
```

```python
# Encode the diagnosis values
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df.iloc[:, 1] = labelencoder.fit_transform(df.iloc[:, 1].values)
```

```python
# display the dataframe
df
```

```python
# Splitting the dataset into independent and dependent dataset
X = df.iloc[:, 2:].values
Y = df.iloc[:, 1].values
```

```python
# Explicit Type conversion for Y to float64
Y = Y.astype('float64')
```

```python
# Spliting datasets into training(75%) and testing(25%)
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25)
```

```python
# Scaling the data(feature scaling)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

Build Logistic Regression Model

```python
# Bluild logistic regression classifier
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, Y_train)
```

```python
# Make use of trained model to make predictions on test data
predictions = classifier.predict(X_test)
```

Performance Evaluation

```python
# Polt confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
cm = confusion_matrix(Y_test, predictions)
sns.heatmap(cm, annot=True)
```

```python
# get accuracy score for model
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, predictions)
```
